```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
volt = np.loadtxt('7-1_data.txt')
```

```python
settings = np.loadtxt('7-1_settings.txt')
settings

charge = settings[0]
finish = settings[1]
dt = settings[2]
dv = settings[3]
charge
```

```python
time = np.linspace(0, finish, len(volt))
```

```python
plt.grid()
plt.plot(time, volt, c='orange', linewidth=0.2)
plt.axvline(charge, linestyle=':', c='black', linewidth=1)

plt.text(charge, volt.max() - 1.5, "Время зарядки: {:.5f} с".format(charge))
plt.text(charge, volt.max() - 1.6, "Время разрядки:  {:.5f} с".format(finish - charge))
plt.xlabel(r'$Время$, $c$',    wrap=True)
plt.ylabel(r'Напряжение, $В$', wrap=True)

plt.title('Процесс заряда и разряда конденсатора в RC цепочке', wrap=True)
plt.savefig('plot.pdf')
```
